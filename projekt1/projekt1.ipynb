## Datorseende Projekt 1
- Programmet kan öppna, spara, rita linjer och ramar på bilder. 
- "l" = Rita linje
- "r" = Rita rektangel
- "o" = Öppna en bild som finns i img foldern
- "s" = Spara bilden i img foldern 
- "q" = Stäng av programmet

```python
import cv2 as cv
import numpy as np
import os
import tkinter as tk
from tkinter import simpledialog

# DrawLine ritar en linje från start positionen rills att man släpper musknappen
 
def drawLine(event, x, y, flags, param):
    global drawing, startPos, img
    # Ta en kopia av den nuvarande bilden
    imgCopy = img.copy() 
    # När man trycker ner den vänstra musknappen så skall man spara start positionen och drawing sätts till True
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        startPos = (x,y)
    # När man rör på musknappen och drawing är True så uppdateras den nuvarande bilden 
    if event == cv.EVENT_MOUSEMOVE and drawing:
        cv.line(imgCopy,(startPos[0],startPos[1]),(x,y),(0,150,0),5)
        cv.imshow("img",imgCopy)
    # När man släpper musknappen så sätts drawing till False och 
    if event == cv.EVENT_LBUTTONUP:
        drawing = False
        cv.line(img,(startPos[0],startPos[1]),(x,y),(0,150,0),5)
        cv.imshow("img",img)

# DrawRect ritar en ram från start positionen tills att man släpper musknappen

def drawRect(event, x, y, flags, param):
    global drawing, startPos, img
    # Ta en kopia av den nuvarande bilden
    imgCopy = img.copy()
    # När man trycker ner den vänstra musknappen så skall man spara start positionen och drawing sätts till True
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        startPos = (x,y)
    # När man rör på musknappen och drawing är True så uppdateras den nuvarande bilden     
    if event == cv.EVENT_MOUSEMOVE and drawing:
        cv.rectangle(imgCopy, pt1=(startPos[0],startPos[1]), pt2=(x, y),color=(0,255,255),thickness=5)
        cv.imshow("img",imgCopy)
    # När man släpper musknappen så sätts drawing till False och     
    if event == cv.EVENT_LBUTTONUP:
        drawing = False
        cv.rectangle(img, pt1=(startPos[0],startPos[1]), pt2=(x, y),color=(0,255,255),thickness=5)
        cv.imshow("img",img)
        
# FileName funktionen frågar efter ett namn och öppnar en bild med det namnet som finns i img foldern

def fileName():
    global img
    # ROOT är tkinters root fönster som man gömmer med withdraw för vi behöver ändast ett simpelt promptfönster
    ROOT.withdraw()
    # om man satt fel filnamn så används message variabeln för att meddela användaren att filen inte hittades.
    message = ""
    while True:
        # Fråga efter filnamn
        USER_INP = simpledialog.askstring(title="Filname", prompt=message+"Open file: \n\n"+"\n".join(os.listdir("img/")))
        # om användaren  tryckt på cancel  knappen så är USER_INP == None
        if USER_INP == None:
            return False
        # Om filnamnet finns i img foldern
        elif USER_INP in os.listdir("img/"):
            # Läs in bilden till img variabeln
            img = cv.imread("img/"+USER_INP)
            # Visa bilden
            cv.imshow("img",img)
            return True
        else:
            # om man satt in fel filnamn meddela användaren om detta
            message = "Wrong filename!!\n"

# SaveFile frågar efter ett filnamn och sparar bilden i img folder

def saveFile():
    global img
    # ROOT är tkinters root fönster som man gömmer med withdraw för vi behöver ändast ett simpelt promptfönster
    ROOT.withdraw()
    # Fråga efter filnamn
    USER_INP = simpledialog.askstring(title="Filname", prompt="Save as: ")
    # Skriver bilden till en fil med dett givna namnet
    cv.imwrite("img/"+USER_INP+".jpg",img)


# Global variables        
startPos = (0,0)
drawing = False   
img = np.zeros(shape=[512, 512, 3], dtype=np.uint8)  
ROOT = tk.Tk()
isFileValid = fileName()

# Om användaren tryckt på Cancel knappen så är isFileValid == False
# Om filen har hittats så är isFileValid == True
if isFileValid:
    # Loopa igenom tills användaren trycker på "q"
    while True:
        cv.imshow("img",img)
        if cv.waitKey() == ord("q"):
            break
        elif cv.waitKey() == ord("o"):
            fileName()
        elif cv.waitKey() == ord("l"):
            cv.setMouseCallback('img', drawLine)
        elif cv.waitKey() == ord("r"):
            cv.setMouseCallback('img', drawRect)
        elif cv.waitKey() == ord("s"):
            saveFile()
        elif cv.waitKey() == ord("q"):
            cv.destroyAllWindows() 
            break

cv.destroyAllWindows() 


```
